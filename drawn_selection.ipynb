```python
import json

def read_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    time = []
    hv = []
    for key, value in data.items():
        time.append(value["time"])
        hv.append(value["HV"])
    return time, hv
```

```python
import matplotlib.pyplot as plt

time_1, hv_1 = read_data(r"Result_selection\time_objective\Surrogate\conus_urban_easy_s3.json")
time_2, hv_2 = read_data(r"Result_selection\time_objective\NSGAII\conus_urban_easy_s3.json")
plt.plot(time_1, hv_1, label="Surrogate")
plt.plot(time_2, hv_2, label="NSGAII")
plt.legend()
```

```python
from utils.utils import cal_hv
import numpy as np
import json

def read_hv(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    
    objective = []
    for item in data["test_result"]:
        objective.append([item["normal_reject"], item["normal_cost"]])
    return cal_hv(np.array(objective), np.array([1, 1]))
```
