```python
from utils.utils import cal_hv
import json
import numpy as np

def cal_HV(data_path, ref_point):
    nsga_ii = "Pareto_objective_history/NSGAII/" + data_path
    moea_d = "Pareto_objective_history/MOEAD/" + data_path
    with open(nsga_ii, "r") as f:
        nsga_ii_data = json.load(f)
    with open(moea_d, "r") as f:
        moead_data = json.load(f)
    nsga_ii_hv = []
    moead_hv = []
    for i in range(100):
        nsga_ii_Pareto = []
        moead_Pareto = []
        for value in nsga_ii_data[str(i+1)]:
            nsga_ii_Pareto.append([value["obj1"], value["obj2"]])
        for value in moead_data[str(i+1)]:
            moead_Pareto.append([value["obj1"], value["obj2"]])
        nsga_ii_hv.append(cal_hv(np.array(nsga_ii_Pareto), np.array(ref_point)))
        moead_hv.append(cal_hv(np.array(moead_Pareto), np.array(ref_point)))
    return nsga_ii_hv, moead_hv


```

```python
import matplotlib.pyplot as plt
nsga_ii_hv, moead_hv = cal_HV("1.json", [1, 1])
plt.plot(nsga_ii_hv, label="NSGAII")
plt.plot(moead_hv, label="MOEAD")
plt.legend()
print(nsga_ii_hv)
print(moead_hv)
```
