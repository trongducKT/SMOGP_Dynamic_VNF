```python
import json
import matplotlib.pyplot as plt

with open(r'surrogate_value_1.json', 'r') as f:
    data = json.load(f)

predict_objective_1 = data['predict_objectives_1']
predict_objective_2 = data['predict_objectives_2']
extract_objective_1 = data['extract_objectives_1']
extract_objective_2 = data['extract_objectives_2']

plt.scatter(predict_objective_1, predict_objective_2, c='r', label='predict')
plt.scatter(extract_objective_1, extract_objective_2, c='b', label='extract', alpha=0.9)
plt.legend()
plt.plot()
```

```python

import numpy as np
different_objective_1 = np.abs(np.array(predict_objective_1) - np.array(extract_objective_1))/np.array(extract_objective_1)*100
different_objective_2 = np.abs(np.array(predict_objective_2) - np.array(extract_objective_2))/np.array(extract_objective_2)*100
plt.scatter(different_objective_1, different_objective_2)

print(np.mean(different_objective_1), np.std(different_objective_1))
print(np.mean(different_objective_2), np.std(different_objective_2))
plt.plot()
```
